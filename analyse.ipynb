```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open('grid.txt', 'r')
phi = np.loadtxt(f).reshape(99, 99)
plt.plot(np.arange(99) ,phi[:][55])

```

```python
fig, ax = plt.subplots(nrows=12, ncols = 1, figsize=(5, 16))
for i, n in enumerate([10, 30, 50, 70, 80, 100, 130, 160, 180, 200, 1000, 10000, 100000]):
    f = open('grid' + str(n) + 'steps.txt', 'r')
    phi = np.loadtxt(f).reshape(99, 99)
    ax[i].plot(np.arange(99) ,phi[:][55])
```
